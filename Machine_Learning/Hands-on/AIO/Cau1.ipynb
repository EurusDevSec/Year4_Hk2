1. Viết hàm tính độ đo F1-Score thường được sử dụng để đánh
giá mô hình phân loại.

• Precision =
TP
TP + FP

• Recall =
TP
TP + FN
• F1-score = 2 ∗

Precision ∗ Recall
Precision + Recall
• Input: hàm nhận 3 giá trị tp, fp, fn
• Output: trả về và in ra kết quả của Precision, Recall, và F1-
score

```python
def evaluate_f1_components(tp=2, fp=3, fn=4):
    if (type(tp) != "int" and type(fp) != "int" and type(fn) != "int"):

        print("fn must be int")
        exit()
    try:
        if tp >=0 and fp >=0 and fn >=0:
            

            prec = tp/(tp + fp)
            Recall = tp/(fp + fn)
            f1_score = 2 * (prec * Recall)/(prec + Recall)
        else:
            print("tp and fp and fn must be greater than or equal zero")
            exit()
    except ZeroDivisionError:
        print("khong duoc nhap so 0 ")
    pass

evaluate_f1_components()
```
