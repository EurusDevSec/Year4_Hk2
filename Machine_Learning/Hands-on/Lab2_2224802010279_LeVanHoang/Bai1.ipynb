## 1. Hàm Sigmoid

### Công thức

$$ \sigma(x) = \frac{1}{1 + e^{-x}} $$

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def sigmoid(x):
    return 1/ (1+ np.exp(-x))
```

```python
# Ví dụ kiểm tra

x = np.linspace(-10, 10,100)
plt.plot(x, sigmoid(x))
plt.title("Hàm Sigmoid")
plt.grid()
plt.show()
```

### **Ý nghĩa và Công dụng**
- **Đặc điểm:** Đưa ra giá trị đầu vào về khoảng (0,1). Đồ thị có hình chữ S.
- **Công dụng:** 
    - Thường được dùng ở lớp đầu ra (output layer)  cho các bài toán `phân loại nhị phân` (binary classification) vì giá trị đầu ra có thể giải thích là xác xuất 
- **Hạn chế:** Dễ gặp vấn đề "biến mất đạo hàm" khi giá trị đầu vào quá lớn hoặc quá nhỏ, làm chậm quá trình học của mạng sâu

## **2. Hàm Tank (Hyberbolic Tangent)**

### Công thức

$$ \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

```python
def tanh(x):
    return np.tanh(x)
```

```python
# Vi du kiem thu
x = np.linspace(-10,10,100)
plt.plot(x, tanh(x))
plt.title("Hầm tanh")
plt.grid()
plt.show()
```

### Ý nghĩa và Công dụng

- **Đặc điểm**: Đưa giá trị đầu vào về khoảng $(-1, 1)$. Đồ thị cũng có hình chữ S nhưng đi qua gốc tọa độ $(0,0)$.
- **Công dụng**:
  - Thường tốt hơn Sigmoid khi dùng ở các lớp ẩn (hidden layers) vì dữ liệu đầu ra được chuẩn hóa quanh 0 (zero-centered), giúp việc tối ưu hóa dễ dàng hơn.
- **Hạn chế**: Vẫn gặp vấn đề biến mất đạo hàm (vanishing gradient) giống như Sigmoid.

## 3. Hàm ReLU (Rectified Linear Unit)

### Công thức

$$ f(x) = \max(0, x) $$

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-10, 10,100)
plt.plot(x, relu(x))
plt.title("Hàm ReLU")
plt.grid()
plt.show()
```


### Ý nghĩa và Công dụng

- **Đặc điểm**: Nếu $x > 0$, hàm trả về chính nó; nếu $x \le 0$, hàm trả về 0. Tuyến tính ở phần dương.
- **Công dụng**:
  - Là hàm kích hoạt phổ biến nhất hiện nay cho các lớp ẩn trong mạng nơ-ron sâu (Deep Neural Networks).
  - **Giảm thiểu vấn đề biến mất đạo hàm**: Đạo hàm bằng 1 khi $x > 0$.
  - **Tính toán hiệu quả**: Chỉ cần so sánh với 0, không cần tính toán hàm mũ phức tạp.
- **Hạn chế**: Vấn đề "Dying ReLU" - các nơ-ron có thể bị "chết" (luôn trả về 0) nếu trọng số cập nhật sao cho đầu vào luôn âm.

| **Sigmoid** | $(0, 1)$       | Output layer phân loại nhị phân | Dễ hiểu xác suất                         | Vanishing gradient, không zero-centered |
